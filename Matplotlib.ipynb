# Импорт библиотек

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

# Подготовим данные

Данные взяты из google trends по запросам 'Python' и 'C++' за последние 5 лет(2018-2023)

```python
py = []
with open('python.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        py.append(int(row[0]))
```

```python
c = []
with open('c++.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        c.append(int(row[0]))
```

```python
py[:20]
```

# Простейший график

```python
x = np.arange(len(py))

y_py = np.array(py)
y_c = np.array(c)
```

```python
#Построение графика
plt.plot(x, y_py)
```

# Улучшаем визуализацию

Добавим еще одну функцию на график

```python
plt.plot(x, y_py)
plt.plot(x, y_c)
```

Теперь давайте подпишем наши функции и уберем лишнее

```python
plt.plot(x, y_py, label='python')
plt.plot(x, y_c, label='c++')
plt.legend();
```

Уже лучше! Теперь подпишем сам график и оси

```python
plt.plot(x, y_py, label='python')
plt.plot(x, y_c, label='c++')

plt.xlabel('Время')
plt.ylabel('Количество запросов')
plt.title('Количество запросов за 5 лет')

plt.legend();
```

Есть проблемка

```python
x_range = np.linspace(0, len(py), 6)
x_labels = ['2018', '2019', '2020', '2021', '2022', '2023']

plt.plot(x, y_py, label='python')
plt.plot(x, y_c, label='c++')

plt.xticks(x_range, x_labels)

plt.xlabel('Время')
plt.ylabel('Количество запросов')
plt.title('Количество запросов за 5 лет')

plt.legend();
```

А что если нужно добавить сетку?

```python
x_range = np.linspace(0, len(py), 6)
x_labels = ['2018', '2019', '2020', '2021', '2022', '2023']

plt.plot(x, y_py, label='python')
plt.plot(x, y_c, label='c++')

plt.xticks(x_range, x_labels)
plt.grid()

plt.xlabel('Время')
plt.ylabel('Количество запросов')
plt.title('Количество запросов за 5 лет')

plt.legend();
```

А если хочется сделать график больше?

```python
x_range = np.linspace(0, len(py), 6)
x_labels = ['2018', '2019', '2020', '2021', '2022', '2023']

fig = plt.figure(figsize=(12, 4))

plt.plot(x, y_py, label='python')
plt.plot(x, y_c, label='c++')

plt.xticks(x_range, x_labels)
plt.grid()

plt.xlabel('Время')
plt.ylabel('Количество запросов')
plt.title('Количество запросов за 5 лет')

plt.legend();
```

# Добавим разнообразия: графики бывают разные

**Точечный график(scatter)**

Подготовим данные: будем использовать рандом
Предположим, нам нужно визуализировать объекты по их географическим координатам(долгота - longitude и широта - latitude), чтобы найти скопления

```python
long = np.random.uniform(-180, 180, 100)
lat = np.random.uniform(-90, 90, 100)
```

```python
plt.scatter(long, lat)
plt.xlabel('Широта')
plt.ylabel('Долгота')
```

Пусть наши объекты будут разного рамера. Для этого нужно добавить еще один вектор с данными

```python
size = np.random.randint(10, 80, 100)
```

```python
plt.scatter(long, lat, s = size)
plt.xlabel('Широта')
plt.ylabel('Долгота');
```

Отлично, у нас уже есть 3 характеристики каждого объекта! Добавим еще одну - население(обозначим цветом)

```python
population = np.random.randint(10000, 1000000, 100)
```

Красиво, но непонятно. Добавим специальную шкалу

```python
plt.scatter(long, lat, s = size, c = population)
plt.xlabel('Широта')
plt.ylabel('Долгота');
plt.colorbar(label='Население');
```

**Круговая диаграмма(pie)**

Сгенерируем данные опроса, где люди(1000) отвечали на разные вопросы. В данном случае, нас интересует только факт наличия некоторого заболевания(например, ветрянка)

```python
types = ['YES', 'NO', 'NO']

answers = [np.random.choice(types) for i in range(1000)]
```

Визуализируем их ответы

```python
vals = [answers.count('YES'), answers.count('NO')]
```

```python
plt.pie(vals, labels = types[:2]);
```

Сделаем немного понятнее

```python
dist = [0, 0.05]
plt.pie(vals, labels = types[:2], explode = dist, autopct='%.2f');
```

**Задание 1**

Сделайте две области в одном графике. В каждой из них изобразить две разные функции(например, квадрат;куб и тангенc;синус)

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1)
fig = plt.figure(figsize=(6, 4))
x = np.arange(-10, 10, 0.2)
y = np.sin(x)
z = np.cos(x)

axes[0].plot(x, y, label='sin')
axes[0].plot(x, z, label='cos')
axes[0].grid()
axes[0].legend()

y = np.power(x, 2)
z = np.power(x, 3)
axes[1].plot(x, y, label='pow2')
axes[1].plot(x, z, label='pow3')
axes[1].grid()
axes[1].legend();
```

**Задание 3**

Прогноз погоды: постройте график температуры(дневной и ночной) в вашем городе на неделю вперед

```python
day = np.array([15, 17, 18, 15, 12, 7, 6])
night = np.array([8, 13, 11, 10, 7, 4, 3])

x_labels = ['пн', 'вт', 'ср', 'чт', 'пн', 'сб', 'вс']
x_range = np.linspace(0, len(x_labels), 7)
plt.xticks(x_range, x_labels)
plt.xlabel('Дни недели')
plt.ylabel('Температура')

plt.plot(x_range, day, label='День', color='orange')
plt.plot(x_range, night, label='Ночь', color='blue')
plt.grid()
plt.legend();
```
