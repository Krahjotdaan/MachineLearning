## 1.Описательные статистики

### 1.1 Пример с котиками

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
kotiki = np.array([1,3,11,12,14,14,18,18,18])
```

```python
plt.bar(np.arange(len(kotiki)),kotiki)
```

### 1.2 Средние значения
#### Мода

```python
Counter (kotiki).most_common()[0][0]
from scipy import stats
stats.mode(kotiki)
```

#### Медиана

```python
np.median(kotiki)
```

#### Среднее (арифметическое)

```python
kotiki.mean()
```

### 1.3 Изменчивость
#### Минимум и максимум

```python
kotiki.min(), kotiki.max()
```

#### Дисперсия
$$D = \frac{1}{N}\sum^N_{i=1}(x_i - \mathbb{E}X)^2$$

$$где, \mathbb{E}X - среднее  X $$

```python
((kotiki - kotiki.mean())**2).sum()/len(kotiki)
```

```python
np.var(kotiki)
```
