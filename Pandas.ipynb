## Шаг 11. Агрегирование данных и групповые операции

Одна из важнейших операций в работе с данными - это разбиение набора данных на группы и применение функций к каждой группе

Этот механизм состоит из трех операций
- разделение
- применение
- объединение

### 11.1 Группировки

```python
# Найти среднюю оценку каждого признака для каждого возраста
df.groupby('Occupation').mean()
```

```python
df.groupby('Occupation')['Quality of Sleep'].mean()
```

```python
df['Quality of Sleep'].groupby(df['Occupation']).mean()
```

### 12.1 Агрегирование

```python
# напишите свою функцию например для вычисления разности максимального и минимального значения списка
def delta(nums):
  return nums.max() - nums.min()
```

```python
delta(df['Quality of Sleep'].groupby(df['Occupation']))
```

```python
df['Quality of Sleep'].groupby(df['Occupation']).agg(delta)
```
