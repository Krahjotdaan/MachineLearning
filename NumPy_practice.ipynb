Ниже считывается файл с минимальными температурами по дням в период с 01.01.2003 по 25.09.2005 в г. Москва

#### Задание 3.

Вам дан вектор a = [1,2,5,10]

Превратите данный вектор в [0,4,0,20]

в одну строчку. Да, нужно угадать формулу.

```python
a = np.array([1, 2, 5, 10])
a = np.array(list(a[i] * 0 if i % 2 == 0 else a[i] * 2 for i in range(len(a))))
a
```
